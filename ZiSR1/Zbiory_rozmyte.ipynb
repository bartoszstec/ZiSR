```python
import math

class FuzzySet:
    """Abstract base class for a generic fuzzy set."""
    def __init__(self, name: str):
        self.name = name

    def _membership(self, x: float) -> float:
        """Abstract method to calculate the membership degree for a given x."""
        raise NotImplementedError("Subclasses must implement _membership method")

    def membership(self, x: float) -> float:
        """Returns the membership degree for a given x, ensuring it's within [0, 1]."""
        mu = self._membership(x)
        return max(0.0, min(1.0, mu)) # Ensure membership is between 0 and 1

    def __str__(self):
        return f"Fuzzy Set: {self.name}"


class TriangularFuzzySet(FuzzySet):
    """Represents a triangular fuzzy set defined by three points (a, b, c)."""
    def __init__(self, name: str, a: float, b: float, c: float):
        super().__init__(name)
        if not (a <= b <= c):
            raise ValueError("For a triangular fuzzy set, a <= b <= c must hold.")
        self.a = a
        self.b = b
        self.c = c

    def _membership(self, x: float) -> float:
        if x <= self.a or x >= self.c:
            return 0.0
        elif self.a < x <= self.b:
            return (x - self.a) / (self.b - self.a)
        elif self.b < x < self.c:
            return (self.c - x) / (self.c - self.b)
        return 0.0 # Should not be reached

    def __str__(self):
        return f"Triangular Fuzzy Set: {self.name} (a={self.a}, b={self.b}, c={self.c})"


class TrapezoidalFuzzySet(FuzzySet):
    """Represents a trapezoidal fuzzy set defined by four points (a, b, c, d)."""
    def __init__(self, name: str, a: float, b: float, c: float, d: float):
        super().__init__(name)
        if not (a <= b <= c <= d):
            raise ValueError("For a trapezoidal fuzzy set, a <= b <= c <= d must hold.")
        self.a = a
        self.b = b
        self.c = c
        self.d = d

    def _membership(self, x: float) -> float:
        if x <= self.a or x >= self.d:
            return 0.0
        elif self.a < x <= self.b:
            return (x - self.a) / (self.b - self.a)
        elif self.b < x <= self.c:
            return 1.0
        elif self.c < x < self.d:
            return (self.d - x) / (self.d - self.c)
        return 0.0 # Should not be reached

    def __str__(self):
        return f"Trapezoidal Fuzzy Set: {self.name} (a={self.a}, b={self.b}, c={self.c}, d={self.d})"


class GaussianFuzzySet(FuzzySet):
    """Represents a Gaussian fuzzy set defined by a mean and standard deviation."""
    def __init__(self, name: str, mean: float, std_dev: float):
        super().__init__(name)
        if std_dev <= 0:
            raise ValueError("Standard deviation must be positive.")
        self.mean = mean
        self.std_dev = std_dev

    def _membership(self, x: float) -> float:
        return math.exp(-0.5 * ((x - self.mean) / self.std_dev) ** 2)

    def __str__(self):
        return f"Gaussian Fuzzy Set: {self.name} (mean={self.mean}, std_dev={self.std_dev})"


# --- Przykłady użycia ---
print("\n--- Przykłady użycia ---")

# Zbiór trójkątny: 'Mały'
triangular_set = TriangularFuzzySet("Mały", a=0, b=5, c=10)
print(triangular_set)
print(f"Przynależność 3 do 'Mały': {triangular_set.membership(3):.2f}")
print(f"Przynależność 5 do 'Mały': {triangular_set.membership(5):.2f}")
print(f"Przynależność 8 do 'Mały': {triangular_set.membership(8):.2f}")
print(f"Przynależność 12 do 'Mały': {triangular_set.membership(12):.2f}")

# Zbiór trapezowy: 'Średni'
trapezoidal_set = TrapezoidalFuzzySet("Średni", a=5, b=10, c=15, d=20)
print(f"\n{trapezoidal_set}")
print(f"Przynależność 7 do 'Średni': {trapezoidal_set.membership(7):.2f}")
print(f"Przynależność 12 do 'Średni': {trapezoidal_set.membership(12):.2f}")
print(f"Przynależność 18 do 'Średni': {trapezoidal_set.membership(18):.2f}")
print(f"Przynależność 22 do 'Średni': {trapezoidal_set.membership(22):.2f}")

# Zbiór gaussowski: 'Duży'
gaussian_set = GaussianFuzzySet("Duży", mean=25, std_dev=3)
print(f"\n{gaussian_set}")
print(f"Przynależność 22 do 'Duży': {gaussian_set.membership(22):.2f}")
print(f"Przynależność 25 do 'Duży': {gaussian_set.membership(25):.2f}")
print(f"Przynależność 28 do 'Duży': {gaussian_set.membership(28):.2f}")
print(f"Przynależność 35 do 'Duży': {gaussian_set.membership(35):.2f}")
```
